```python
import pandas as pd
train_df = pd.read_csv("/content/train_df_MEGNET.csv").drop(columns=["Unnamed: 0"])
test_df = pd.read_csv("/content/test_df_MEGNET.csv").drop(columns=["Unnamed: 0"])
test_df
```

## Add labels

```python
import tarfile
fname = "/content/IDAO-2022/data/dichalcogenides_public.tar.gz"
if fname.endswith("tar.gz"):
    tar = tarfile.open(fname, "r:gz")
    tar.extractall()
    tar.close()
elif fname.endswith("tar"):
    tar = tarfile.open(fname, "r:")
    tar.extractall()
    tar.close()
```

```python
labels = pd.read_csv("/content/dichalcogenides_public/targets.csv")
```

```python
features = train_df.drop(columns=["id", "formula", "structure", "composition"]).to_numpy()
targets  = labels['band_gap'].to_numpy()
```

```python
test_df = test_df.drop(columns=["id", "formula", "structure", "composition"])
```

## Fit CatBoost

```python
from catboost import CatBoostRegressor
```

```python
best_params = {'loss_function': 'MAE',
               'learning_rate': 0.0747810942910428,
               'l2_leaf_reg': 0.012316070330352875,
               'colsample_bylevel': 0.09026019271098612,
               'depth': 11,
               'boosting_type': 'Plain',
               'bootstrap_type': 'MVS',
               'min_data_in_leaf': 16,
               'one_hot_max_size': 4,
               'n_estimators': 8000,
               'random_seed': 42}

CATBOOST = CatBoostRegressor(**best_params)
CATBOOST.fit(features, targets, verbose=0)
```

```python
preds = CATBOOST.predict(test_df.to_numpy())
```

# Submit

```python
test_df = pd.read_csv("/content/test_df_MEGNET.csv").drop(columns=["Unnamed: 0"])
ids = test_df['id']
```

```python
submission = pd.DataFrame({'id': ids, 'predictions': preds})
submission
```

```python
submission.to_csv('submission.csv')
```
