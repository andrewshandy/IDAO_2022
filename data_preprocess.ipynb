```python
from megnet.models import MEGNetModel
from pymatgen.ext.matproj import MPRester
from megnet.models import MEGNetModel
from megnet.data.crystal import CrystalGraph
import pandas as pd
import numpy as np
import tensorflow as tf
from pymatgen.core.structure import Structure


def energy_within_threshold(prediction, target):
    # compute absolute error on energy per system.
    # then count the no. of systems where max energy error is < 0.02.
    e_thresh = 0.02
    error_energy = tf.math.abs(target - prediction)

    success = tf.math.count_nonzero(error_energy < e_thresh)
    total = tf.size(target)
    return success / tf.cast(total, tf.int64)

def prepare_model(cutoff, lr):
    nfeat_bond = 10
    r_cutoff = cutoff
    gaussian_centers = np.linspace(0, r_cutoff + 1, nfeat_bond)
    gaussian_width = 0.8
    
    return MEGNetModel(
        graph_converter=CrystalGraph(cutoff=r_cutoff),
        centers=gaussian_centers,
        width=gaussian_width,
        loss=["MAE"],
        npass=2,
        lr=lr,
        metrics=energy_within_threshold
)
    
model = prepare_model(4, 2e-4)
model.load_weights("callback/val_mae_00779_0.040351.hdf5")


```

```python
import json, os
from tqdm import tqdm
def get_data(path_list, root):
    dataset = []
    for file_name in tqdm(path_list):  
        file_path = os.path.join(root, file_name + ".json")
        f = open(file_path) # open file
        dict_file = json.load(f)
        mol = Structure.from_dict(dict_file)
        preds = model.predict_structure(mol)
        if len(preds) > 1:
            print(preds)
        preds = preds[0]
        dataset.append([file_name, mol, mol.formula,
                        preds])
    return pd.DataFrame(dataset, columns=["id", "structure", "formula", "MEGnet"])
```

```python
import xgboost as xgb
labels = pd.read_csv("/content/dichalcogenides_public/targets.csv")

train_root = "/content/dichalcogenides_public/structures"
train_df = get_data(labels["_id"], train_root)
train_df
```

```python
from matminer.featurizers.structure import DensityFeatures, StructuralComplexity, StructureComposition
from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.conversions import CompositionToOxidComposition
from matminer.featurizers.composition import OxidationStates
from matminer.featurizers.composition import ElementProperty



def featurize(df):
    df = StrToComposition().featurize_dataframe(df, "formula")

    ep_feat = ElementProperty.from_preset(preset_name="magpie")
    feat = StructuralComplexity()
    df = feat.featurize_dataframe(df, "structure")
    # df = StructureComposition().featurize_dataframe(df, "structure")

    df_feat = DensityFeatures()
    df = df_feat.featurize_dataframe(df, "structure")  # input the structure column to the featurizer

    df = ep_feat.featurize_dataframe(df, col_id="composition")

    
    # df = CompositionToOxidComposition().featurize_dataframe(df, "composition")
    # os_feat = OxidationStates()
    # df = os_feat.featurize_dataframe(df, "composition_oxid")
    return df

```

```python
train_df = featurize(train_df)
train_df
```

```python
def get_test_data(root):
    dataset = []
    for file_name in tqdm(os.listdir(root)):  
        file_path = os.path.join(root, file_name)
        f = open(file_path) # open file
        dict_file = json.load(f)
        mol = Structure.from_dict(dict_file)
        preds = model.predict_structure(mol)[0]
        dataset.append([file_name, mol, mol.formula,
                        preds])
    return pd.DataFrame(dataset, columns=["id", "structure", "formula", "MEGnet"])

test_df = get_test_data("/content/dichalcogenides_private/structures")
test_df
```
